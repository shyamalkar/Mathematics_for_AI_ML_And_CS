***`Simple coding for better understanding`***

```python
# P(A), P(B), P(A and B)
P_A = 0.5
P_B = 0.4
P_A_and_B = 0.2

P_A_or_B = P_A + P_B - P_A_and_B

print(P_A_or_B)
# addition rule → overlapping events handle korte use hoy (double counting avoid)
```

***`Probability Multiplical Tule (A and B)`***

```python
# independent events
P_A = 0.5
P_B = 0.4

P_A_and_B = P_A * P_B

print(P_A_and_B)
# multiplication rule → joint probability (duita event ekshathe ghotar chance)
```
